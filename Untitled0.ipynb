```python
%%html
<!DOCTYPE html>
<html lang="zh-TW">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>ç§‘å­¸æœŸåˆŠRSSèšåˆå™¨ - ç¥ç¶“ç§‘å­¸èˆ‡ç”Ÿç‰©é†«å­¸ç ”ç©¶</title>
    <script src="https://cdn.tailwindcss.com"></script>
    <link href="https://fonts.googleapis.com/css2?family=Inter:wght@400;500;600;700&family=Roboto+Mono:wght@400;500&display=swap" rel="stylesheet">
    <style>
        body {
            font-family: 'Inter', sans-serif;
        }
        .mono {
            font-family: 'Roboto Mono', monospace;
        }
        ::-webkit-scrollbar {
            width: 10px;
        }
        ::-webkit-scrollbar-track {
            background: #f1f5f9;
        }
        ::-webkit-scrollbar-thumb {
            background: #64748b;
            border-radius: 5px;
        }
        ::-webkit-scrollbar-thumb:hover {
            background: #475569;
        }
        .article-card {
            transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
        }
        .article-card:hover {
            transform: translateY(-2px);
        }
        .journal-badge {
            font-size: 0.75rem;
            padding: 0.125rem 0.5rem;
            border-radius: 9999px;
            font-weight: 600;
        }
        .category-filter {
            transition: all 0.2s ease;
        }
        .category-filter:hover {
            transform: scale(1.05);
        }
        .loading-shimmer {
            background: linear-gradient(90deg, #f0f0f0 25%, #e0e0e0 50%, #f0f0f0 75%);
            background-size: 200% 100%;
            animation: shimmer 1.5s infinite;
        }
        @keyframes shimmer {
            0% { background-position: 200% 0; }
            100% { background-position: -200% 0; }
        }
    </style>
</head>
<body class="bg-slate-50 text-slate-900">

    <div class="min-h-screen flex flex-col">
        <!-- é ‚éƒ¨å°èˆª -->
        <header class="bg-white shadow-sm sticky top-0 z-30 border-b border-slate-200">
            <div class="max-w-7xl mx-auto px-4 sm:px-6 lg:px-8 py-4">
                <div class="flex justify-between items-center">
                    <div>
                        <h1 class="text-2xl font-bold text-slate-900">ç§‘å­¸æœŸåˆŠRSSèšåˆå™¨</h1>
                        <p class="text-sm text-slate-600 mt-1">ç¥ç¶“ç§‘å­¸ â€¢ è—¥ç†å­¸ â€¢ è…«ç˜¤å­¸ â€¢ å¤–æ³Œé«”å·¥ç¨‹</p>
                    </div>
                    <div class="flex items-center space-x-4">
                        <button id="refreshAllBtn" class="bg-slate-100 text-slate-700 px-4 py-2 rounded-lg hover:bg-slate-200 focus:outline-none focus:ring-2 focus:ring-offset-2 focus:ring-slate-500 transition flex items-center">
                            <svg class="w-4 h-4 mr-2" fill="none" stroke="currentColor" viewBox="0 0 24 24">
                                <path stroke-linecap="round" stroke-linejoin="round" stroke-width="2" d="M4 4v5h.582m15.356 2A8.001 8.001 0 004.582 9m0 0H9m11 11v-5h-.581m0 0a8.003 8.003 0 01-15.357-2m15.357 2H15"></path>
                            </svg>
                            é‡æ–°æ•´ç†
                        </button>
                        <button id="manageFeedsBtn" class="bg-indigo-600 text-white px-4 py-2 rounded-lg shadow hover:bg-indigo-700 focus:outline-none focus:ring-2 focus:ring-offset-2 focus:ring-indigo-500 transition">
                            ç®¡ç†æœŸåˆŠè¨‚é–±
                        </button>
                    </div>
                </div>

                <!-- åˆ†é¡éæ¿¾å™¨ -->
                <div class="mt-4 flex flex-wrap gap-2">
                    <button data-category="all" class="category-filter px-4 py-1.5 rounded-full bg-indigo-600 text-white text-sm font-medium">
                        å…¨éƒ¨é¡¯ç¤º
                    </button>
                    <button data-category="neuroscience" class="category-filter px-4 py-1.5 rounded-full bg-slate-200 text-slate-700 text-sm font-medium hover:bg-slate-300">
                        ç¥ç¶“ç§‘å­¸
                    </button>
                    <button data-category="pharmacology" class="category-filter px-4 py-1.5 rounded-full bg-slate-200 text-slate-700 text-sm font-medium hover:bg-slate-300">
                        è—¥ç†å­¸
                    </button>
                    <button data-category="oncology" class="category-filter px-4 py-1.5 rounded-full bg-slate-200 text-slate-700 text-sm font-medium hover:bg-slate-300">
                        è…«ç˜¤å­¸
                    </button>
                    <button data-category="bioengineering" class="category-filter px-4 py-1.5 rounded-full bg-slate-200 text-slate-700 text-sm font-medium hover:bg-slate-300">
                        ç”Ÿç‰©å·¥ç¨‹
                    </button>
                    <button data-category="ml-ai" class="category-filter px-4 py-1.5 rounded-full bg-slate-200 text-slate-700 text-sm font-medium hover:bg-slate-300">
                        ML/AIæ‡‰ç”¨
                    </button>
                </div>
            </div>
        </header>

        <!-- ä¸»è¦å…§å®¹å€ -->
        <main class="flex-grow max-w-7xl w-full mx-auto px-4 sm:px-6 lg:px-8 py-8">
            <!-- çµ±è¨ˆè³‡è¨Š -->
            <div id="stats-container" class="grid grid-cols-1 md:grid-cols-4 gap-4 mb-8">
                <div class="bg-white p-4 rounded-lg shadow-sm border border-slate-200">
                    <h3 class="text-sm font-medium text-slate-600">ä»Šæ—¥æ–‡ç« </h3>
                    <p class="text-2xl font-bold text-slate-900 mt-1" id="todayCount">0</p>
                </div>
                <div class="bg-white p-4 rounded-lg shadow-sm border border-slate-200">
                    <h3 class="text-sm font-medium text-slate-600">æœ¬é€±æ–‡ç« </h3>
                    <p class="text-2xl font-bold text-slate-900 mt-1" id="weekCount">0</p>
                </div>
                <div class="bg-white p-4 rounded-lg shadow-sm border border-slate-200">
                    <h3 class="text-sm font-medium text-slate-600">è¨‚é–±æœŸåˆŠ</h3>
                    <p class="text-2xl font-bold text-slate-900 mt-1" id="feedCount">0</p>
                </div>
                <div class="bg-white p-4 rounded-lg shadow-sm border border-slate-200">
                    <h3 class="text-sm font-medium text-slate-600">ä¸Šæ¬¡æ›´æ–°</h3>
                    <p class="text-sm font-semibold text-slate-900 mt-1 mono" id="lastUpdate">--:--</p>
                </div>
            </div>

            <!-- è¼‰å…¥æŒ‡ç¤ºå™¨ -->
            <div id="loading-indicator" class="text-center py-12">
                <svg class="animate-spin h-8 w-8 text-indigo-600 mx-auto" xmlns="http://www.w3.org/2000/svg" fill="none" viewBox="0 0 24 24">
                    <circle class="opacity-25" cx="12" cy="12" r="10" stroke="currentColor" stroke-width="4"></circle>
                    <path class="opacity-75" fill="currentColor" d="M4 12a8 8 0 018-8V0C5.373 0 0 5.373 0 12h4zm2 5.291A7.962 7.962 0 014 12H0c0 3.042 1.135 5.824 3 7.938l3-2.647z"></path>
                </svg>
                <p class="mt-4 text-slate-600">æ­£åœ¨è¼‰å…¥æœŸåˆŠæ–‡ç« ...</p>
            </div>

            <!-- ç©ºç‹€æ…‹ -->
            <div id="empty-state" class="hidden text-center py-12 bg-white rounded-lg shadow border border-slate-200">
                <svg xmlns="http://www.w3.org/2000/svg" class="mx-auto h-12 w-12 text-slate-400" fill="none" viewBox="0 0 24 24" stroke="currentColor" stroke-width="1">
                    <path stroke-linecap="round" stroke-linejoin="round" d="M12 6.253v13m0-13C10.832 5.477 9.246 5 7.5 5S4.168 5.477 3 6.253v13C4.168 18.477 5.754 18 7.5 18s3.332.477 4.5 1.253m0-13C13.168 5.477 14.754 5 16.5 5c1.747 0 3.332.477 4.5 1.253v13C19.832 18.477 18.247 18 16.5 18c-1.746 0-3.332.477-4.5 1.253"></path>
                </svg>
                <h3 class="mt-2 text-sm font-medium text-slate-900">å°šæœªè¨‚é–±ä»»ä½•æœŸåˆŠ</h3>
                <p class="mt-1 text-sm text-slate-500">é–‹å§‹è¨‚é–±æ‚¨ç ”ç©¶é ˜åŸŸçš„é ‚ç´šæœŸåˆŠ</p>
                <div class="mt-6">
                    <button id="addFirstFeedBtn" type="button" class="inline-flex items-center px-4 py-2 border border-transparent shadow-sm text-sm font-medium rounded-md text-white bg-indigo-600 hover:bg-indigo-700 focus:outline-none focus:ring-2 focus:ring-offset-2 focus:ring-indigo-500">
                        <svg class="-ml-1 mr-2 h-5 w-5" xmlns="http://www.w3.org/2000/svg" viewBox="0 0 20 20" fill="currentColor" aria-hidden="true">
                            <path fill-rule="evenodd" d="M10 3a1 1 0 011 1v5h5a1 1 0 110 2h-5v5a1 1 0 11-2 0v-5H4a1 1 0 110-2h5V4a1 1 0 011-1z" clip-rule="evenodd" />
                        </svg>
                        æ–°å¢æœŸåˆŠ
                    </button>
                </div>
            </div>

            <!-- æ–‡ç« å®¹å™¨ -->
            <div id="articles-container" class="space-y-4">
                <!-- æ–‡ç« å°‡åœ¨æ­¤è™•å‹•æ…‹è¼‰å…¥ -->
            </div>
        </main>
    </div>

    <!-- æœŸåˆŠç®¡ç†æ¨¡æ…‹æ¡† -->
    <div id="modal" class="hidden fixed inset-0 bg-slate-900 bg-opacity-50 overflow-y-auto h-full w-full z-40">
        <div class="relative top-20 mx-auto p-5 border w-full max-w-2xl shadow-xl rounded-lg bg-white">
            <div class="mt-3">
                <div class="flex justify-between items-center mb-4">
                    <h3 class="text-lg leading-6 font-medium text-slate-900">ç®¡ç†æœŸåˆŠè¨‚é–±</h3>
                    <button id="closeModalBtn" class="text-slate-400 hover:text-slate-600 text-2xl">&times;</button>
                </div>

                <!-- å¿«é€Ÿæ–°å¢æ¨è–¦æœŸåˆŠ -->
                <div class="mb-6">
                    <h4 class="text-md font-medium text-slate-800 mb-3">æ¨è–¦æœŸåˆŠå¿«é€Ÿè¨‚é–±</h4>
                    <div class="grid grid-cols-2 gap-2">
                        <button class="quick-add-btn text-left p-2 bg-slate-100 rounded hover:bg-slate-200 text-sm" data-url="https://www.nature.com/nrn.rss" data-name="Nature Reviews Neuroscience" data-category="neuroscience">
                            ğŸ§  Nature Reviews Neuroscience
                        </button>
                        <button class="quick-add-btn text-left p-2 bg-slate-100 rounded hover:bg-slate-200 text-sm" data-url="https://www.nature.com/neuro.rss" data-name="Nature Neuroscience" data-category="neuroscience">
                            ğŸ”¬ Nature Neuroscience
                        </button>
                        <button class="quick-add-btn text-left p-2 bg-slate-100 rounded hover:bg-slate-200 text-sm" data-url="https://www.science.org/action/showFeed?type=etoc&feed=rss&jc=science" data-name="Science" data-category="all">
                            ğŸ“š Science
                        </button>
                        <button class="quick-add-btn text-left p-2 bg-slate-100 rounded hover:bg-slate-200 text-sm" data-url="https://www.cell.com/cell/rss" data-name="Cell" data-category="all">
                            ğŸ§¬ Cell
                        </button>
                        <button class="quick-add-btn text-left p-2 bg-slate-100 rounded hover:bg-slate-200 text-sm" data-url="https://www.nature.com/nrd.rss" data-name="Nature Reviews Drug Discovery" data-category="pharmacology">
                            ğŸ’Š Nature Reviews Drug Discovery
                        </button>
                        <button class="quick-add-btn text-left p-2 bg-slate-100 rounded hover:bg-slate-200 text-sm" data-url="https://www.nature.com/nrc.rss" data-name="Nature Reviews Cancer" data-category="oncology">
                            ğŸ—ï¸ Nature Reviews Cancer
                        </button>
                        <button class="quick-add-btn text-left p-2 bg-slate-100 rounded hover:bg-slate-200 text-sm" data-url="https://www.nature.com/natcancer.rss" data-name="Nature Cancer" data-category="oncology">
                            ğŸ”¬ Nature Cancer
                        </button>
                        <button class="quick-add-btn text-left p-2 bg-slate-100 rounded hover:bg-slate-200 text-sm" data-url="https://www.nature.com/nbt.rss" data-name="Nature Biotechnology" data-category="bioengineering">
                            ğŸ§ª Nature Biotechnology
                        </button>
                    </div>
                </div>

                <!-- æ‰‹å‹•æ–°å¢æœŸåˆŠ -->
                <form id="addFeedForm" class="space-y-4 mb-6">
                    <div>
                        <label for="feedUrl" class="block text-sm font-medium text-slate-700">æ‰‹å‹•æ–°å¢RSSè¨‚é–±ç¶²å€</label>
                        <div class="mt-1 flex rounded-md shadow-sm">
                            <input type="url" name="feedUrl" id="feedUrl" class="flex-1 block w-full rounded-l-md border-slate-300 focus:ring-indigo-500 focus:border-indigo-500" placeholder="https://example.com/feed.rss" required>
                            <select id="feedCategory" class="border-slate-300 focus:ring-indigo-500 focus:border-indigo-500">
                                <option value="neuroscience">ç¥ç¶“ç§‘å­¸</option>
                                <option value="pharmacology">è—¥ç†å­¸</option>
                                <option value="oncology">è…«ç˜¤å­¸</option>
                                <option value="bioengineering">ç”Ÿç‰©å·¥ç¨‹</option>
                                <option value="ml-ai">ML/AIæ‡‰ç”¨</option>
                                <option value="other">å…¶ä»–</option>
                            </select>
                            <button type="submit" class="inline-flex items-center px-3 rounded-r-md border border-l-0 border-slate-300 bg-slate-50 text-slate-500 hover:bg-slate-100">æ–°å¢</button>
                        </div>
                        <p class="mt-2 text-sm text-slate-500">ä¾‹å¦‚ï¼šhttps://www.nature.com/nrn.rss</p>
                    </div>
                </form>

                <!-- ç›®å‰è¨‚é–±æ¸…å–® -->
                <div class="mt-6">
                    <h4 class="text-md font-medium text-slate-800">ç›®å‰è¨‚é–±çš„æœŸåˆŠ</h4>
                    <div id="feed-list-container" class="mt-2 space-y-2 max-h-60 overflow-y-auto">
                        <!-- æœŸåˆŠæ¸…å–®å°‡åœ¨æ­¤è™•å‹•æ…‹è¼‰å…¥ -->
                    </div>
                </div>
            </div>
        </div>
    </div>

    <!-- è¨Šæ¯æ¨¡æ…‹æ¡† -->
    <div id="messageModal" class="hidden fixed inset-0 bg-slate-900 bg-opacity-50 overflow-y-auto h-full w-full z-50">
        <div class="relative top-20 mx-auto p-5 border w-full max-w-sm shadow-xl rounded-lg bg-white">
            <div class="mt-3 text-center">
                <h3 id="messageModalTitle" class="text-lg leading-6 font-medium text-slate-900"></h3>
                <div class="mt-2 px-7 py-3">
                    <p id="messageModalText" class="text-sm text-slate-600"></p>
                </div>
                <div class="items-center px-4 py-3">
                    <button id="messageModalCloseBtn" class="px-4 py-2 bg-indigo-600 text-white text-base font-medium rounded-md w-full shadow-sm hover:bg-indigo-700 focus:outline-none focus:ring-2 focus:ring-indigo-500">
                        é—œé–‰
                    </button>
                </div>
            </div>
        </div>
    </div>

    <script>
        // === å…¨åŸŸè®Šæ•¸èˆ‡é…ç½® ===
        const CORS_PROXY = "https://api.allorigins.win/get?url=";
        let currentFeeds = [];
        let allArticles = [];
        let currentCategory = 'all';

        // æœŸåˆŠåˆ†é¡å°æ‡‰çš„é¡è‰²
        const categoryColors = {
            neuroscience: 'bg-purple-100 text-purple-800',
            pharmacology: 'bg-green-100 text-green-800',
            oncology: 'bg-red-100 text-red-800',
            bioengineering: 'bg-blue-100 text-blue-800',
            'ml-ai': 'bg-yellow-100 text-yellow-800',
            other: 'bg-slate-100 text-slate-800'
        };

        // === DOM å…ƒç´  ===
        const manageFeedsBtn = document.getElementById('manageFeedsBtn');
        const addFirstFeedBtn = document.getElementById('addFirstFeedBtn');
        const closeModalBtn = document.getElementById('closeModalBtn');
        const modal = document.getElementById('modal');
        const addFeedForm = document.getElementById('addFeedForm');
        const feedUrlInput = document.getElementById('feedUrl');
        const feedCategorySelect = document.getElementById('feedCategory');
        const feedListContainer = document.getElementById('feed-list-container');
        const articlesContainer = document.getElementById('articles-container');
        const loadingIndicator = document.getElementById('loading-indicator');
        const emptyState = document.getElementById('empty-state');
        const messageModal = document.getElementById('messageModal');
        const messageModalTitle = document.getElementById('messageModalTitle');
        const messageModalText = document.getElementById('messageModalText');
        const messageModalCloseBtn = document.getElementById('messageModalCloseBtn');
        const refreshAllBtn = document.getElementById('refreshAllBtn');

        // === å·¥å…·å‡½æ•¸ ===
        function showMessage(title, text, type = 'info') {
            messageModalTitle.textContent = title;
            messageModalText.textContent = text;
            messageModal.classList.remove('hidden');
        }

        function getFromLocalStorage(key) {
            try {
                return JSON.parse(localStorage.getItem(key) || '[]');
            } catch {
                return [];
            }
        }

        function saveToLocalStorage(key, data) {
            localStorage.setItem(key, JSON.stringify(data));
        }

        function generateId() {
            return Date.now().toString(36) + Math.random().toString(36).substr(2);
        }

        // === RSS è§£æ ===
        async function fetchAndParseRSS(feed) {
            try {
                const response = await fetch(`${CORS_PROXY}${encodeURIComponent(feed.url)}`);
                if (!response.ok) throw new Error(`Network response was not ok`);

                const data = await response.json();
                const text = data.contents;

                const parser = new DOMParser();
                const xmlDoc = parser.parseFromString(text, "text/xml");

                const errorNode = xmlDoc.querySelector("parsererror");
                if (errorNode) throw new Error("Failed to parse RSS feed.");

                const items = Array.from(xmlDoc.querySelectorAll("item, entry"));
                return items.map(item => {
                    const title = item.querySelector("title")?.textContent || "[ç„¡æ¨™é¡Œ]";
                    const link = item.querySelector("link")?.getAttribute('href') ||
                                item.querySelector("link")?.textContent || "#";
                    const description = item.querySelector("description")?.textContent ||
                                      item.querySelector("summary")?.textContent || "";
                    const pubDateStr = item.querySelector("pubDate")?.textContent ||
                                     item.querySelector("published")?.textContent ||
                                     item.querySelector("updated")?.textContent;
                    const pubDate = pubDateStr ? new Date(pubDateStr) : new Date();

                    // è§£æä½œè€…ï¼ˆå¦‚æœæœ‰ï¼‰
                    const author = item.querySelector("author")?.textContent ||
                                 item.querySelector("dc\\:creator")?.textContent || "";

                    return {
                        title,
                        link,
                        description,
                        pubDate,
                        feedUrl: feed.url,
                        feedName: feed.name,
                        category: feed.category,
                        author
                    };
                });
            } catch (error) {
                console.error(`Error fetching feed ${feed.url}:`, error);
                return [];
            }
        }

        // === UI æ¸²æŸ“ ===
        function renderArticles(articles) {
            const filteredArticles = currentCategory === 'all'
                ? articles
                : articles.filter(a => a.category === currentCategory);

            articlesContainer.innerHTML = '';

            if (filteredArticles.length === 0) {
                articlesContainer.innerHTML = `
                    <div class="text-center py-12">
                        <p class="text-slate-500">æ­¤åˆ†é¡ç›®å‰æ²’æœ‰æ–‡ç« </p>
                    </div>
                `;
                return;
            }

            // æ ¹æ“šæ—¥æœŸæ’åº
            filteredArticles.sort((a, b) => b.pubDate - a.pubDate);

            filteredArticles.forEach(article => {
                const articleEl = document.createElement('div');
                articleEl.className = 'article-card bg-white p-6 rounded-lg shadow-sm border border-slate-200 hover:shadow-md';

                // æ¸…ç†æè¿°ä¸­çš„HTML
                const tempDiv = document.createElement('div');
                tempDiv.innerHTML = article.description;
                const cleanDescription = tempDiv.textContent.trim();
                const snippet = cleanDescription.substring(0, 300) + (cleanDescription.length > 300 ? '...' : '');

                // æ ¼å¼åŒ–æ—¥æœŸ
                const formattedDate = new Intl.DateTimeFormat('zh-TW', {
                    year: 'numeric',
                    month: 'numeric',
                    day: 'numeric',
                    hour: '2-digit',
                    minute: '2-digit'
                }).format(article.pubDate);

                articleEl.innerHTML = `
                    <div class="flex items-start justify-between mb-3">
                        <span class="journal-badge ${categoryColors[article.category || 'other']}">
                            ${article.feedName}
                        </span>
                        <time class="text-xs text-slate-500 mono">${formattedDate}</time>
                    </div>
                    <h2 class="text-lg font-bold mb-2 text-slate-900">
                        <a href="${article.link}" target="_blank" rel="noopener noreferrer"
                           class="hover:text-indigo-600 transition-colors">
                            ${article.title}
                        </a>
                    </h2>
                    ${article.author ? `<p class="text-sm text-slate-600 mb-2">ä½œè€…ï¼š${article.author}</p>` : ''}
                    <div class="text-sm text-slate-700 leading-relaxed">${snippet}</div>
                    <div class="mt-4 flex items-center space-x-4">
                        <a href="${article.link}" target="_blank" rel="noopener noreferrer"
                           class="text-sm text-indigo-600 hover:text-indigo-800 font-medium">
                            é–±è®€å…¨æ–‡ â†’
                        </a>
                    </div>
                `;
                articlesContainer.appendChild(articleEl);
            });

            updateStats(allArticles);
        }

        function renderFeedList() {
            feedListContainer.innerHTML = '';
            if (currentFeeds.length === 0) {
                feedListContainer.innerHTML = `<p class="text-sm text-slate-500">å°šæœªè¨‚é–±ä»»ä½•æœŸåˆŠ</p>`;
                return;
            }

            currentFeeds.forEach(feed => {
                const feedEl = document.createElement('div');
                feedEl.className = 'flex justify-between items-center bg-slate-50 p-3 rounded-lg';
                feedEl.innerHTML = `
                    <div class="flex-1">
                        <span class="font-medium text-slate-800">${feed.name}</span>
                        <span class="ml-2 text-xs ${categoryColors[feed.category]} px-2 py-1 rounded-full">
                            ${getCategoryName(feed.category)}
                        </span>
                        <p class="text-xs text-slate-500 mt-1 mono">${feed.url}</p>
                    </div>
                    <button data-id="${feed.id}" class="remove-feed-btn text-red-500 hover:text-red-700 font-bold ml-4 text-xl">
                        Ã—
                    </button>
                `;
                feedListContainer.appendChild(feedEl);
            });
        }

        function getCategoryName(category) {
            const names = {
                neuroscience: 'ç¥ç¶“ç§‘å­¸',
                pharmacology: 'è—¥ç†å­¸',
                oncology: 'è…«ç˜¤å­¸',
                bioengineering: 'ç”Ÿç‰©å·¥ç¨‹',
                'ml-ai': 'ML/AIæ‡‰ç”¨',
                other: 'å…¶ä»–'
            };
            return names[category] || 'å…¶ä»–';
        }

        function updateStats(articles) {
            const now = new Date();
            const today = new Date(now.getFullYear(), now.getMonth(), now.getDate());
            const weekAgo = new Date(today.getTime() - 7 * 24 * 60 * 60 * 1000);

            const todayArticles = articles.filter(a => a.pubDate >= today);
            const weekArticles = articles.filter(a => a.pubDate >= weekAgo);

            document.getElementById('todayCount').textContent = todayArticles.length;
            document.getElementById('weekCount').textContent = weekArticles.length;
            document.getElementById('feedCount').textContent = currentFeeds.length;
            document.getElementById('lastUpdate').textContent = new Intl.DateTimeFormat('zh-TW', {
                hour: '2-digit',
                minute: '2-digit',
                second: '2-digit'
            }).format(now);
        }

        // === äº‹ä»¶è™•ç† ===
        function toggleModal() {
            modal.classList.toggle('hidden');
        }

        manageFeedsBtn.addEventListener('click', toggleModal);
        addFirstFeedBtn.addEventListener('click', toggleModal);
        closeModalBtn.addEventListener('click', toggleModal);
        messageModalCloseBtn.addEventListener('click', () => messageModal.classList.add('hidden'));

        // åˆ†é¡éæ¿¾å™¨
        document.querySelectorAll('.category-filter').forEach(btn => {
            btn.addEventListener('click', () => {
                document.querySelectorAll('.category-filter').forEach(b => {
                    b.classList.remove('bg-indigo-600', 'text-white');
                    b.classList.add('bg-slate-200', 'text-slate-700');
                });
                btn.classList.remove('bg-slate-200', 'text-slate-700');
                btn.classList.add('bg-indigo-600', 'text-white');

                currentCategory = btn.dataset.category;
                renderArticles(allArticles);
            });
        });

        // å¿«é€Ÿæ–°å¢æœŸåˆŠ
        document.querySelectorAll('.quick-add-btn').forEach(btn => {
            btn.addEventListener('click', () => {
                const feed = {
                    id: generateId(),
                    url: btn.dataset.url,
                    name: btn.dataset.name,
                    category: btn.dataset.category,
                    addedAt: new Date().toISOString()
                };

                if (!currentFeeds.find(f => f.url === feed.url)) {
                    currentFeeds.push(feed);
                    saveToLocalStorage('journal_feeds', currentFeeds);
                    renderFeedList();
                    showMessage('æˆåŠŸ', `å·²æ–°å¢ ${feed.name}`);
                    loadAllFeeds(); // é‡æ–°è¼‰å…¥æ–‡ç« 
                } else {
                    showMessage('æç¤º', 'æ­¤æœŸåˆŠå·²è¨‚é–±');
                }
            });
        });

        // æ‰‹å‹•æ–°å¢æœŸåˆŠ
        addFeedForm.addEventListener('submit', (e) => {
            e.preventDefault();
            const url = feedUrlInput.value.trim();
            const category = feedCategorySelect.value;

            if (url) {
                const feed = {
                    id: generateId(),
                    url: url,
                    name: new URL(url).hostname,
                    category: category,
                    addedAt: new Date().toISOString()
                };

                if (!currentFeeds.find(f => f.url === feed.url)) {
                    currentFeeds.push(feed);
                    saveToLocalStorage('journal_feeds', currentFeeds);
                    renderFeedList();
                    feedUrlInput.value = '';
                    showMessage('æˆåŠŸ', 'å·²æ–°å¢æœŸåˆŠè¨‚é–±');
                    loadAllFeeds();
                } else {
                    showMessage('æç¤º', 'æ­¤æœŸåˆŠå·²è¨‚é–±');
                }
            }
        });

        // ç§»é™¤æœŸåˆŠ
        feedListContainer.addEventListener('click', (e) => {
            if (e.target.classList.contains('remove-feed-btn')) {
                const feedId = e.target.dataset.id;
                currentFeeds = currentFeeds.filter(f => f.id !== feedId);
                saveToLocalStorage('journal_feeds', currentFeeds);
                renderFeedList();
                showMessage('æˆåŠŸ', 'å·²ç§»é™¤æœŸåˆŠè¨‚é–±');
                loadAllFeeds();
            }
        });

        // é‡æ–°æ•´ç†
        refreshAllBtn.addEventListener('click', () => {
            loadAllFeeds();
        });

        // === ä¸»è¦åŠŸèƒ½ ===
        async function loadAllFeeds() {
            if (currentFeeds.length === 0) {
                loadingIndicator.classList.add('hidden');
                emptyState.classList.remove('hidden');
                articlesContainer.innerHTML = '';
                return;
            }

            loadingIndicator.classList.remove('hidden');
            emptyState.classList.add('hidden');
            articlesContainer.innerHTML = '';

            const articlePromises = currentFeeds.map(feed => fetchAndParseRSS(feed));
            const articleArrays = await Promise.all(articlePromises);
            allArticles = articleArrays.flat();

            loadingIndicator.classList.add('hidden');
            renderArticles(allArticles);
        }

        // === åˆå§‹åŒ– ===
        function init() {
            // è¼‰å…¥å·²å„²å­˜çš„æœŸåˆŠ
            currentFeeds = getFromLocalStorage('journal_feeds');
            renderFeedList();
            loadAllFeeds();
        }

        // å•Ÿå‹•æ‡‰ç”¨ç¨‹å¼
        init();
    </script>

</body>
</html>
```
